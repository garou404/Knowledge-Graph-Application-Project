```python
def scrape_elements_v2(url, class_name):
    # Set up the Selenium WebDriver
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")  # Run in headless mode
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
    
    try:
        # Open the webpage
        driver.get(url)
        time.sleep(5)  # Wait for JavaScript to load content

        # Get the page source after rendering
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Find all elements with the given class
        elements = soup.find_all(class_=class_name)

        return [element.get_text(strip=True) for element in elements]
    
    finally:
        driver.quit()  # Close the browser
```

```python
url = "https://universe.leagueoflegends.com/en_GB/region/" 
class_name = "factionName_2meu"

# Parse the page content with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find all elements with the given class
elements = soup.find_all(class_=class_name)
elements = [element.get_text(strip=True) for element in elements]
faction_list = []
for i, element in enumerate(elements, 1):
	faction_list.append(element)
```

```python
dict_champion_region = {
      'Champion': [],
      'Region': []
}

for faction in faction_list:
    faction_url = faction.lower().replace('the ', '').replace(' ', '-').replace('targon', 'mount-targon')
    champions = scrape_elements_v2(url+faction_url+"/", "copy_xxN7")
    for champion in champions:
            
            dict_champion_region['Champion'].append(champion)
            dict_champion_region['Region'].append(faction)
    
    
df_region = pd.DataFrame(data=dict_champion_region)
df_region['Champion'] = df_region.apply(lambda x : x['Champion'].replace(x['Region'], '').replace('Freljord', ''), axis=1)
```

```python
def scrape_table(url, class_name):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return None
    
    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the table with the given class
    table = soup.find('table', class_=class_name)
    
    if not table:
        print("Table not found")
        return None
    
    # Extract table rows
    rows = table.find_all('tr')
    data = []
    
    for row in rows:
        cols = row.find_all(['th', 'td'])
        data.append([col.get_text(strip=True) for col in cols])
    
    # Convert to DataFrame
    df = pd.DataFrame(data)
    return df
```
